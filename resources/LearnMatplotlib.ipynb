```python
##Subplots and Legends
```

```python
days = [1,2,3,4,5]
original_price = [10,20,25,30,40]
predicted_price = [8,21,25,28,38]
```

```python
plt.plot(days , original_price , 'b-o' ,label = 'Original Price');
plt.plot(days , predicted_price , 'r-o' ,label = 'Predicted Price');
plt.title('Plots with Legends');
plt.xlabel('WeekDays');
plt.ylabel('Price');
plt.legend();
```

```python
##Subplots
```

```python
plt.subplot(3 , 1 , 1)
plt.plot(days , original_price , 'b-o' ,label = 'Original Price');
plt.title('Original prices over time');
plt.xlabel('WeekDays');
plt.ylabel('Price');
plt.legend();

plt.subplot(3 , 1 , 2)
plt.plot(days , predicted_price , 'r-o' ,label = 'Predicted Price');
plt.title('Predicted prices over time');
plt.xlabel('WeekDays');
plt.ylabel('Price');
plt.legend();

plt.subplot(3 , 1 , 3)
price_diff = [o-p for o , p in zip(original_price , predicted_price)]
plt.bar(days , price_diff , color = 'purple' ,label = 'Price Difference');
plt.title('Price Difference over time');
plt.xlabel('WeekDays');
plt.ylabel('Price Difference');
plt.legend();
plt.tight_layout()
```

```python
##Save Image
```

```python
plt.plot(days , original_price , 'b-o' ,label = 'Original Price');
plt.plot(days , predicted_price , 'r-o' ,label = 'Predicted Price');
plt.title('Plots with Legends');
plt.xlabel('WeekDays');
plt.ylabel('Price');
plt.legend();
plt.savefig('output.png' , dpi=300)
```
